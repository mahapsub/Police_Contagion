```python
'''
This file generates some starter facts and stores it in starter_facts.krf
Adjust the sample size in percent to get more/less officers

'''


import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')


SAMPLE_SIZE = .002

```

```python
df = pd.read_csv('data_officernetwork.csv')
```

```python
df.columns
```

```python
selected_df = df[['first_name','last_name','rank','race',
    'gender','officer_age','trr_percentile',
   'internal_allegation_percentile','complaint_percentile']]
```

```python
selected_df
```

```python
selected_df['trr_percentile'] = selected_df.trr_percentile.apply(lambda x: 'high' if int(x) > 50 else 'low')
selected_df['internal_allegation_percentile'] = selected_df.internal_allegation_percentile.apply(lambda x: 'high' if int(x) > 50 else 'low')
selected_df['complaint_percentile'] = selected_df.complaint_percentile.apply(lambda x: 'high' if int(x) > 50 else 'low')
selected_df['officer_age'] = selected_df.officer_age.apply(lambda x: 'middle-aged' if int(x) > 0 and int(x) <40 else 'old')
```

```python
selected_df
```

```python
selected_df['full_name'] = selected_df.first_name+' '+selected_df.last_name
```

```python
row_values = selected_df.values
```

```python
sample_rows = row_values[0:int(SAMPLE_SIZE*len(row_values))]
len(sample_rows)
```

```python
sample_rows[0]
```

```python
#making ISA Persons

isa_lst = []
for i in sample_rows:
    tmp_builder = "(isa "
    tmp_builder +=(i[0]+'_'+i[1]+' ')
    tmp_builder +='person)'
    isa_lst.append(tmp_builder)
print(isa_lst)
```

```python
#associating ranks to people
rank_lst = []
for i in sample_rows:
    rank = i[2].replace(" ", "_")
    tmp_builder = "(rank "
    tmp_builder +=(i[0]+'_'+i[1]+' ')
    tmp_builder +='{})'.format(rank)
    rank_lst.append(tmp_builder)
print(rank_lst)
```

```python
#associating Race to people
race_lst = []
for i in sample_rows:
    race = i[3]
    tmp_builder = "(race "
    tmp_builder +=(i[0]+'_'+i[1]+' ')
    tmp_builder +='{})'.format(race)
    race_lst.append(tmp_builder)
print(race_lst)
```

```python
#associating Gender to people
gender_lst = []
for i in sample_rows:
    gender = i[4].replace('M', 'Male')
    gender = gender.replace('F', 'Female')
    tmp_builder = "(gender "
    tmp_builder +=(i[0]+'_'+i[1]+' ')
    tmp_builder +='{})'.format(gender)
    gender_lst.append(tmp_builder)
print(gender_lst)
```

```python
#associating Age to people
age_lst = []
for i in sample_rows:
    age = i[5]
    tmp_builder = "(age "
    tmp_builder +=(i[0]+'_'+i[1]+' ')
    tmp_builder +='{})'.format(age)
    age_lst.append(tmp_builder)
print(age_lst)
```

```python
selected_df
```

```python
#associating TRR to people
trr_lst = []
for i in sample_rows:
    trr = i[6]
    tmp_builder = "(trr_percentile "
    tmp_builder +=(i[0]+'_'+i[1]+' ')
    tmp_builder +='{})'.format(trr)
    trr_lst.append(tmp_builder)
print(trr_lst)
```

```python
#associating internal_allegation to people
alleg_lst = []
for i in sample_rows:
    alegpercent = i[7]
    tmp_builder = "(internal_allegation_percentile "
    tmp_builder +=(i[0]+'_'+i[1]+' ')
    tmp_builder +='{})'.format(alegpercent)
    alleg_lst.append(tmp_builder)
print(alleg_lst)
```

```python
#associating complaint_allegation to people
comp_lst = []
for i in sample_rows:
    compercent = i[8]
    tmp_builder = "(complaint_percentile "
    tmp_builder +=(i[0]+'_'+i[1]+' ')
    tmp_builder +='{})'.format(compercent)
    comp_lst.append(tmp_builder)
print(comp_lst)
```

```python
#associating full_name to people
name_lst = []
for i in sample_rows:
    name = i[9].replace(' ','')
    tmp_builder = "(full_name "
    tmp_builder +=(i[0]+'_'+i[1]+' ')
    tmp_builder +='{})'.format(name)
    name_lst.append(tmp_builder)
print(name_lst)
```

```python
# isa_lst
# rank_lst
# race_lst
# gender_lst
# age_lst
# trr_lst
# alleg_lst
# comp_lst
# name_lst

f = open('starter_facts.krf', 'w')
for i in range(len(isa_lst)):
    f.write("%s\n" % isa_lst[i])
    f.write("%s\n" % rank_lst[i])
    f.write("%s\n" % race_lst[i])
    f.write("%s\n" % gender_lst[i])
    f.write("%s\n" % age_lst[i])
    f.write("%s\n" % trr_lst[i])
    f.write("%s\n" % alleg_lst[i])
    f.write("%s\n" % comp_lst[i])
    f.write("%s\n" % name_lst[i])
f.close()

```
